```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import tree
import pickle
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```
# loading dataset
main_path = '/content/drive/My Drive/Fake news detection/preprocessing/real_and_fake/'
dataset =  pd.read_csv(main_path+'fake_or_real_news_features_original_with_label.csv')
X = dataset[['Article Length','Average Word Length','Count of Numbers',
             'Count of Exclaimation Marks','Count of Adjectives','Word Count',
             'Negative Sentiment','Positive Sentiment','Neutral Sentiment']]
Y = dataset['Label']
```

```
# splitting data into train and test 80/20
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```
len(X_test)
```

```
# creating model 
DT_classifier = tree.DecisionTreeClassifier()

# training the model
DT_classifier.fit(X_train, y_train)

# saving the model file
pickle.dump(DT_classifier, open('/content/drive/My Drive/Fake news detection/Traditional methods/Decision_Trees/Real_and_Fake/DT.sav', 'wb'))
```

```
# test Decision Tree model
pred = DT_classifier.predict(X_test)
```

```
# checking accuracy, precision, recall and F1 scores
print ("Accuracy: ")
print (accuracy_score(y_test, pred))

print ("Precision: ")
print (precision_score(y_test, pred, average="weighted"))

print ("Recall: ")
print (recall_score(y_test, pred, average="weighted"))

print ("F1 Score ")
print (f1_score(y_test, pred, average="weighted"))
```

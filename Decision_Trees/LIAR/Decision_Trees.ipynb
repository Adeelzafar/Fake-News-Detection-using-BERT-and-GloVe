```
import pandas as pd
from sklearn.tree import tree
import pickle
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```
# loading dataset
main_path = '/content/drive/My Drive/Fake news detection/preprocessing/liar/'
dataset =  pd.read_csv(main_path+'liar_train_features.csv')
X = dataset[['Article Length','Average Word Length','Count of Numbers',
             'Count of Exclaimation Marks','Count of Adjectives','Word Count',
             'Negative Sentiment','Positive Sentiment','Neutral Sentiment']]
Y = dataset['Label']
```

```
len([x for x in Y if x == True])
```

```
# creating model 
DT_classifier = tree.DecisionTreeClassifier()

# training the model
DT_classifier.fit(X, Y)

# saving the model file
pickle.dump(DT_classifier, open('/content/drive/My Drive/Fake news detection/Traditional methods/Decision_Trees/LIAR/DT.sav', 'wb'))
```

```
# reading test files
dataset =  pd.read_csv(main_path+'liar_test_validation_features.csv')
test_X = dataset[['Article Length','Average Word Length','Count of Numbers',
             'Count of Exclaimation Marks','Count of Adjectives','Word Count',
             'Negative Sentiment','Positive Sentiment','Neutral Sentiment']]
test_Y = dataset['Label']
```

```
len([x for x in test_Y if x == True])
```

```
# test Decision Tree model
pred = DT_classifier.predict(test_X)
```

```
# checking accuracy, precision, recall and F1 scores
print ("Accuracy: ")
print (accuracy_score(test_Y, pred))

print ("Precision: ")
print (precision_score(test_Y, pred, average="weighted"))

print ("Recall: ")
print (recall_score(test_Y, pred, average="weighted"))

print ("F1 Score ")
print (f1_score(test_Y, pred, average="weighted"))
```
